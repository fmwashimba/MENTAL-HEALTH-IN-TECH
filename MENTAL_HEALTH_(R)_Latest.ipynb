## **PROBLEM STATEMENT**
The tech industry is known for its high-pressure work environments, long hours, and a culture that prioritises productivity and success over personal well-being. These factors can contribute to the development of mental health problems such as anxiety, depression, and burnout. Despite the prevalence of mental health issues in the tech industry, there is still a stigma surrounding mental health, which prevents many employees from seeking help.

To address this issue, a research project on mental health in the tech industry will be conducted. The project aims to understand the prevalence of mental health problems, the barriers to seeking treatment, and the impact of mental health on job performance and productivity. By raising awareness about the importance of mental health, the project seeks to reduce stigma and identify ways in which the tech industry can better support its employees' mental health needs. The ultimate goal of this research project is to improve the overall well-being of tech industry employees and promote a healthier work environment.

1. What is the prevalence of mental health conditions among tech industry employees across different countries and regions?

4. What factors influence an employee's decision to seek treatment for mental health conditions?
5. What is the impact of family history of mental illness on the prevalence of mental health conditions in tech industry employees?
6. Are there differences in the prevalence of mental health conditions between self-employed and traditionally employed tech industry workers?
7. What are the most common barriers to seeking treatment for mental health conditions among tech industry employees, and how can these barriers be addressed? 
8. What are the most effective interventions for addressing mental health conditions in the tech industry, and how can these interventions be implemented in different countries and regions?
9. How does the impact of mental health conditions on work performance vary between employees who have sought treatment and those who have not?
10. Are there significant differences in the prevalence of mental health conditions among tech industry employees living in different states or territories within the United States?

````r
#dealing with the age outlier
age_iqr <- IQR(mentalh_clean$Age)

# Calculate the upper and lower bounds for outliers
upper <- quantile(mentalh_clean$Age, 0.75) + 1.5 * age_iqr
lower <- quantile(mentalh_clean$Age, 0.25) - 1.5 * age_iqr

# Identify outliers in the age variable
outliers <- mentalh_clean$Age[mentalh_clean$Age < lower | mentalh_clean$Age > upper]

# Print the outliers
print(outliers)
````

````r
# Finding out how many outliers there are
length(outliers)
````

````r
# Checking for the mode of age variable
get_mode <- function(x) {
  unique_x <- unique(x)
  unique_x[which.max(tabulate(match(x, unique_x)))]}
````

````r
# Calculate the mode of the Age variable
age_mode <- get_mode(mentalh_clean$Age)
````

````r
# Print the mode of the Age variable
age_mode
````

````r
# Replacing the outliers with the mode since the mean is extreme thanks to the outliers
mentalh_clean$Age <- ifelse(mentalh_clean$Age < lower | mentalh_clean$Age > upper|age_mode,mentalh_clean$Age)
library(graphics)
````

Of the tech industry employees who have sought Mental Health treatment is there any interference with their work performance?

````r
# Filter the data to include only people working in the tech firms and who have sought treatment

techMHI_df <- mentalh_clean %>% filter( sought.treatment.for.MHI== "Yes" & Is.employer.tech.company == "Yes"& MHI.interference.with.work != "NO MIH")

# Extract the "Interference with work" variable from the data frame
techMHI_work <- techMHI_df$MHI.interference.with.work

techMHI_summary <- techMHI_df %>% 
  group_by(MHI.interference.with.work) %>% 
 ## summarize(count = n()) 
  tally(name = "count")


# Create a bar plot of the MHI interference with work variable
ggplot(techMHI_summary, aes(x = MHI.interference.with.work, y = count, fill = MHI.interference.with.work)) +
  geom_bar(stat = "identity", color = "black") +
  labs(title = "Interference of MHI with Work in Tech Firms",
       x = "MHI Interference with Work", y = "Count", fill = "MHI Interference with Work") +
  theme(legend.position = "none")+
  scale_fill_manual(values = c("#003f5c", "#ffa600","#bc5090", "#f95d6a")) +
  my_theme() 

# Majority of the respondents said that their illness affects their work sometimes.
# Those who said often and rarely tied
````

How does the impact of mental health conditions on work performance vary between employees who have sought treatment and those who have not?

````r
# Filtering out NO MHI
impact_MHI_df <- mentalh_clean %>% filter(MHI.interference.with.work != "NO MHI")

# Count the number of observations for each combination of variables
impact_MHI_count <- impact_MHI_df %>% 
  group_by(sought.treatment.for.MHI, MHI.interference.with.work) %>% 
##  summarize(count = n())
  tally(name = "count")

# Create the bar plot
ggplot(impact_MHI_count, aes(x = sought.treatment.for.MHI, y = count, fill = MHI.interference.with.work, weight = count)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = count), position = position_stack(vjust = 0.5)) +  # Add labels to the bars
  labs(title = "Impact of Mental Health Interference with Work", 
       x = "Sought Mental Health Treatment", y = "Count", fill = "Level of Interference with Work") +
  scale_fill_manual(values = c("#003f5c", "#ffa600","#bc5090", "#f95d6a", "#f95d6a")) +
  theme_minimal() +
  my_theme()

# Those who have sought treatment mostly said it affects their work sometimes while those who hadn't, majority said it never affected their work 
````

Are there significant differences in the prevalence of mental health conditions among tech industry employees living in different states or territories within the United States?

````r
# Ensuring our independent variable is stored in 1's and 0's before performing logistic regression
mentalh_encode$sought.treatment.for.MHI <- ifelse(mentalh_encode$sought.treatment.for.MHI == 2, 1, 0)
````

````r
# load the required package
library(caTools)
````

````r
# set the seed for reproducibility
set.seed(123)

# split the data into training and testing sets
split <- sample.split(mentalh_encode$sought.treatment.for.MHI, SplitRatio = 0.7)
train <- subset(mentalh_encode, split == TRUE)
test <- subset(mentalh_encode, split == FALSE)
````

````r
# performing the logistic regression
model <- glm(sought.treatment.for.MHI ~ ., data = train, family = "binomial")
summary(model)
````

````r
# Predicting on the test set
pred <- predict(model, newdata = test, type = "response")

# Converting predicted probabilities to class labels
pred_class <- ifelse(pred > 0.5, 1, 0)

# Creating confusion matrix
conf_mat <- table(pred_class, test$sought.treatment.for.MHI)

# Calculating evaluation metrics
accuracy <- sum(diag(conf_mat))/sum(conf_mat)
precision <- conf_mat[2,2]/sum(conf_mat[,2])
recall <- conf_mat[2,2]/sum(conf_mat[2,])
f1_score <- 2 * (precision * recall) / (precision + recall)

# Printing confusion matrix and evaluation metrics
print(conf_mat)
cat("\nAccuracy: ", round(accuracy, 3))
cat("\nPrecision: ", round(precision, 3))
cat("\nRecall: ", round(recall, 3))
cat("\nF1-Score: ", round(f1_score, 3))
````

The logistic regression model had an accuracy of 76.7%.
